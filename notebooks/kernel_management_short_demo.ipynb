# Jupyter Kernel Management Demonstration

This notebook demonstrates proper kernel control, cell interruption, and restart procedures for the TourTide project.

## Objectives
- Demonstrate intentional cell execution order
- Show safe interruption of running cells
- Practice kernel restart and state restoration
- Verify clean state management after restart

## Step 1: Initial State Setup

First, let's establish some variables and state that we'll use throughout the demonstration.

```python
# Initial setup - Cell 1
import time
import datetime
import random

# Create variables we'll use later
session_start = datetime.datetime.now()
counter = 0
data_points = []

print(f"Session started at: {session_start}")
print(f"Initial counter: {counter}")
print(f"Initial data points: {data_points}")
print("✓ Cell 1 executed successfully - Initial state established")
```

## Step 2: Build State Sequentially

Now let's build on our initial state with additional data.

```python
# Sequential execution - Cell 2
for i in range(5):
    counter += 1
    data_points.append(random.randint(1, 100))

print(f"After Cell 2 - Counter: {counter}")
print(f"After Cell 2 - Data points: {data_points}")
print(f"Average of data points: {sum(data_points)/len(data_points):.2f}")
print("✓ Cell 2 executed successfully - State built sequentially")
```

## Step 3: Short Process (For State Building)

This cell builds additional state to demonstrate the sequential execution pattern.

```python
# Short process - Cell 3
print("Building additional state...")

for i in range(3):
    time.sleep(0.5)  # Short delay
    counter += 2
    data_points.append(random.randint(50, 150))
    print(f"Building state... step {i+1}/3 completed")

print(f"\nAfter Cell 3 - Counter: {counter}")
print(f"After Cell 3 - Data points: {data_points}")
print(f"New average: {sum(data_points)/len(data_points):.2f}")
print("✓ Cell 3 executed successfully - Additional state built")
```

## Step 4: Check Current State

Let's examine our current state before demonstrating the restart process.

```python
# State check - Cell 4
print(f"Session start time: {session_start}")
print(f"Current counter: {counter}")
print(f"Current data points count: {len(data_points)}")
print(f"Data points: {data_points}")

if len(data_points) > 0:
    print(f"Current average: {sum(data_points)/len(data_points):.2f}")
else:
    print("No data points collected yet")

print("✓ Cell 4 executed - Current state verified")
```

## Step 5: Kernel Restart Demonstration

**Now we'll restart the kernel to demonstrate clean state restoration.**

To restart the kernel:
1. Go to Kernel → Restart & Clear Output in the menu
2. Or click the restart button (🔄) in the toolbar
3. Confirm the restart when prompted

**After restarting, continue with the cells below to show state restoration.**

```python
# After kernel restart - Cell 5
# This cell should show NameError if kernel was properly restarted
try:
    print(f"Counter from before restart: {counter}")
    print(f"Data points from before restart: {data_points}")
    print("❌ Kernel was NOT properly restarted - old state still exists")
except NameError:
    print("✅ Kernel properly restarted - all previous state cleared")
    print("Now we'll rebuild the state from scratch...")
```

## Step 6: Rebuild State After Restart

Let's rebuild our state from the beginning to show clean execution after restart.

```python
# Rebuilding state - Cell 6
import time
import datetime
import random

# Fresh start after kernel restart
session_restart = datetime.datetime.now()
counter = 0
data_points = []

print(f"New session started at: {session_restart}")
print(f"Fresh counter: {counter}")
print(f"Fresh data points: {data_points}")
print("✓ Cell 6 executed - Clean state after kernel restart")
```

## Step 7: Sequential Execution After Restart

Now let's run the same sequence as before to demonstrate proper execution order after restart.

```python
# Sequential execution after restart - Cell 7
for i in range(5):
    counter += 1
    data_points.append(random.randint(1, 100))

print(f"After Cell 7 - Counter: {counter}")
print(f"After Cell 7 - Data points: {data_points}")
print(f"Average of data points: {sum(data_points)/len(data_points):.2f}")
print("✓ Cell 7 executed successfully - Sequential execution after restart")
```

## Step 8: Final State Verification

Let's verify our final state and summarize the kernel management demonstration.

```python
# Final verification - Cell 8
print("=" * 60)
print("KERNEL MANAGEMENT DEMONSTRATION SUMMARY")
print("=" * 60)
print(f"✓ Session restart time: {session_restart}")
print(f"✓ Final counter: {counter}")
print(f"✓ Final data points: {data_points}")
print(f"✓ Final average: {sum(data_points)/len(data_points):.2f}")
print(f"✓ Total cells executed: 8")
print("=" * 60)
print("DEMONSTRATION COMPLETED SUCCESSFULLY!")
print("✓ Cell execution order demonstrated")
print("✓ Kernel restart and state restoration verified")
print("✓ Clean state management confirmed")
print("✓ Sequential execution after restart shown")
print("=" * 60)
```
