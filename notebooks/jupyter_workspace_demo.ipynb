## System Information Check

```python
# Check Python version and current working directory
import sys
import os
from datetime import datetime

print(f"Python version: {sys.version}")
print(f"Current working directory: {os.getcwd()}")
print(f"Notebook execution time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```

## Simple Python Operations Test

```python
# Test basic Python functionality
import math

# Basic calculations
numbers = list(range(1, 11))
sum_result = sum(numbers)
mean_result = sum_result / len(numbers)
sqrt_result = math.sqrt(25)

print(f"Numbers: {numbers}")
print(f"Sum of numbers 1-10: {sum_result}")
print(f"Mean of numbers 1-10: {mean_result}")
print(f"Square root of 25: {sqrt_result}")
```

## File System Operations

```python
# Demonstrate file system awareness
import pathlib

# Get current notebook path
current_path = pathlib.Path.cwd()
project_root = current_path.parent

print(f"Current directory: {current_path}")
print(f"Project root: {project_root}")
print(f"Parent directory contents: {list(project_root.iterdir())[:5]}")  # Show first 5 items
```
